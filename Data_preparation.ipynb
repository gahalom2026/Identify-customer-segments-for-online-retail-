```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df = df[df['Quantity'] > 0]
```

```python
df = df[df['UnitPrice'] > 0]
```

```python
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
```

```python
customer_data = df.groupby("CustomerID").agg({
    "Quantity":"sum",
    "TotalPrice":"sum",
    "InvoiceNo":"count"
})
```

```python
customer_data.columns = ["TotalQuantity",
                         "TotalSpent",
                         "TotalOrders"]
```
